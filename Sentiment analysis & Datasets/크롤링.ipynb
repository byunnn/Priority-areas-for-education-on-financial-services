<작동 방식>
1. 사용자 입력(페이지 수, 검색어, 검색 방식, 시작 날짜, 끝 날짜) 
2. 네이버 뉴스 홈에 등록된 기사인지 확인하고, 맞으면 해당 url을 메모장으로 저장
3. 웹크롤링 결과 저장 ( 리스트 -> 메모장으로 저장 -> 메모장을 csv로 불러와서 -> 최종 엑셀로 저장)


```python
import requests    #크롤링 라이브러리
import re
from bs4 import BeautifulSoup
import pandas as pd   #CSV 파일 불러오는 라이브러리
from datetime import datetime   #최종 저장 파일 이름을 현시각으로 저장할 라이브러리

RESULT_PATH = "C:/Users/lovej/Desktop/"
now = datetime.now()
```

```python
#pip install fake-useragent
```

```python
maxpage = input("최대 출력할 페이지 수 : ")
query = input("검색 키워드 (디지털): ")
s_date = input("시작날짜 입력(2021.07.01): ")
e_date = input("끝 날짜 입력(2021.07.31): ")

s_from = s_date.replace(".", "")
e_to = e_date.replace(".", "")
page = 1
maxpage_t = (int(maxpage)-1)*10+1
file_name="C:/Users/lovej/Desktop/contents_text.txt"
f = open(file_name, 'w', encoding='utf-8')
```

```python
from fake_useragent import UserAgent
while page < maxpage_t:
        
        print(page)
        
        url = "https://search.naver.com/search.naver?where=news&query=" + query + "&sort=0&ds=" + s_date + "&de=" + e_date + "&nso=so%3Ar%2Cp%3Afrom" + s_from + "to" + e_to + "%2Ca%3A&start=" + str(page)
        ua = UserAgent()
        headers = {
            "User-Agent":
            ua.random
            }
        
        req = requests.get(url, headers=headers)
        print(url)
        cont = req.content
        soup = BeautifulSoup(cont, "html.parser")
        
        for urls in soup.find_all("a", class_="info", target="_blank"):
            try:
                if urls["href"].startswith("https://news.naver.com"):
                    print(urls["href"])
                    n_url = urls["href"]
                    #news_detail = get_news(urls["href"])
                    news_detail = []
                    
                    breq = requests.get(n_url, headers=headers)
                    #print(breq)
                    bsoup = BeautifulSoup(breq.content, "html.parser")
                    
                    title = bsoup.select("h3#articleTitle")[0].text
                    news_detail.append(title)  #index = 0
                    
                    pdate = bsoup.select('.t11')[0].get_text()[:11]
                    news_detail.append(pdate)  #index = 1
                    
                    print(news_detail[1])
                    #print(bsoup.find_all("div", class_="_article_body_contents"))
                    
                    _text = bsoup.find_all("div", class_="_article_body_contents")[0].get_text().replace('\n'," ")
                    btext = _text.replace("// flash 오류를 우회하기 위한 함수 추가 function _flash_removeCallback() {}", "")
                    news_detail.append(btext.strip())  #index = 2
                    
                    #print(news_detail[2])
                    
                    news_detail.append(n_url)   #index = 3
                    
                    #print(news_detail[3])
                    
                    pcompany = bsoup.select("#footer address")[0].a.get_text()
                    news_detail.append(pcompany)   #index = 4
                    
                    f.write("{}\t{}\t{}\t{}\t{}\n".format(news_detail[1], news_detail[4], news_detail[0], news_detail[2],news_detail[3])) # new style
                    #print(news_detail)
            except Exception as e:
                print(e)
                continue
        page += 10
```

```python
# 엑셀로 만들기

data = pd.read_csv('C:/Users/lovej/Desktop/contents_text.txt', sep='\t',header=None, error_bad_lines=False, engine='python')
data.columns = ["years", "company", "title", "contents", "link"]
print(data)
    
xlsx_outputFileName = '%s-%s-%s %s시 %s분 %s초 result.xlsx' % (now.year, now.month, now.day, now.hour, now.minute, now.second)
data.to_excel(RESULT_PATH+xlsx_outputFileName, encoding='utf-8')
```
