### <연습문제>

- #### 주어진 숫자 리스트의 평균을 구하는 함수를 정의하시오.

```python
def _avg(list):
    sum=0
    for i in list:
        sum+=i
    avg=sum/len(list)
    return avg
```

```python
print(_avg([5,5]))
```

```python
_avg([2,3])
```

- #### 두개의 숫자를 입력받아 사칙연산 결과를 출력하는 함수를 정의하시오.

```python
def operation(a,b):
    sum = a+b
    sub = a-b
    mul = a*b
    div = a/b
    print("가 :", sum, "감 :", sub, "승 :", mul, "제 :", div)
    
    
```

```python
operation(12,6)
```
